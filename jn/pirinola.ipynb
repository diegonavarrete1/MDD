# Pirinola
<div> <img src="https://www.hagomitarea.com/wp-content/uploads/2018/11/perinola2.jpg" alt="Pirinola" style="width: 300px;"/></div> 

Juego en el que dos o más jugadores hacen girar por turnos un cilindo hexagonal desde el mango que tiene en la parte superior.

Las caras del hexágono tienen las siguientes inscripciones:
* Toma 1
* Toma 2
* Toma todos
* Pon 1
* Pon 2
* Todos ponen

indicando la acción que debe realizar el jugador cuando la cara correspondiente queda a la vista por arriba.

Al iniciar el juego cada participante coloca en una canasta, que se encuentra sobre la mesa donde se hará girar la pirinola, una cantidad previamente convenida de objetos en disputa (por ejemplo 5 virtual picafresas) del total del que disponen para participar en el juego (ej. 10 virtual picafresas).

Se determina la forma de seleccionar quién inicia el juego (las más alta, la más baja, la que llegó primero, la dueña de la pirinola, etc.) y la siguiente en tirar es quien se encuentre a la derecha de la última tiradora.

Cuando una jugadora se queda sin virtual picafresas ha perido. 

El juego finaliza cuando:
<ol style="list-style-type: upper-alpha">
    <li>el resulado de girar la pirinola es <B>Toma todo</B> </li>
    <li>solo una jugadora se queda con objetos (virtual picafresas)</li>
</ol>

Se puede acordar que después de un **Toma todo** el juego puede continuar sempre y cuando **Todos ponen** nuevamente como antes iniciar nuevamente el juego. En este caso el juego finaliza cuando solo una participante se queda con todos los objetos en juego.

</ol>


## La pirinola
Las características que definen a la pirinola son:
* las caras (Toma 1, Toma 2, Pon 1, Pon 2, Toma todos,Todos ponen)
* la probabilidad que salga cada una de las caras, en una pirinola honesta cada cara tiene la misma probabilidad (1/6)

Las acciones que puede realizar una pirinola
* girar la pirinola, el resultado de esta acción es la cara que queda en la parte superior de la pirinola una vez que para de girar
***
## Clase pirinola
```python
class pirinola(object):
    def __init__(self, ponderadores = None):
        self.set_caras()
        self.set_ponderadores(ponderadores)
    def set_caras(self):
        self.__caras = ("Pon 1", "Pon 2", "Toma 1", "Toma 2", "Toma todo", "Todos ponen")
    def set_ponderadores(self, ponderadores):
        self.__ponderadores = ponderadores
    def get_caras(self):
        return self.__caras
    def get_ponderadores(self):
        return self.__ponderadores
    def girar(self):
        if self.get_ponderadores == None:
            from random import choice
            return choice(self.get_caras())
        else:
            from random import choices
            return choices(self.get_caras(),weights=self.get_ponderadores(), k=1)[0]
```

## Los judadores
Cada jugador tiene las características:
* nombre, puede ser un número aleatorio, un número consecutivo o el nombre de un jugador 
* cantidad de objetos para poner en juego
Las acciones que puede realizar el jugador son:
* recibir objetos (jellybeans o virtual picafresas)
* dar objetos (jellybeans o virtual picafresas). No puede dar más objetos de los que tiene.
***

## La canasta
Las cualidades que definen a la canasta son:
* La cantidad de objetos en juego que almacena (cantidad de jellybeans o picafresas).

Las acciones que pueden realizar una canasta es:
* responder a la pregunta si esta vacía o no
* agregar n objetos a la canasta
* extraer n objetos a la canasta

***
## Clase canasta
```python
class canasta(object):
    def __init__(self):
        self.set_contenido(0)
    def set_contenido(self, jellybeans):
        self.__jellybeans = jellybeans
    def get_contenido(self):
        return self.__jellybeans
    def vacia(self):
        if self.get_contenido() == 0:
            return True
        else:
            False
    def agregar(self, n):
        self.set_contenido(self.get_contenido() + n )
    def extraer(self, n):
        if n <= self.get_contenido():
            self.set_contenido( self.get_contenido() - n )
        else:
            self.set_contenido(0)
    def __str__(self):
        return f"La canasta tiene {self.get_contenido()} picafresas"
            
```



````python
from pirinola import canasta

canasta1 = canasta()
print(canasta1.vacia())
print(canasta1)
canasta1.agregar(15)
print(canasta1)
canasta1.extraer(3)
print(canasta1)
````

## La mesa de juego
La mesa es el punto de encuentro de todos los elementos previos, sobre la mesa esta la canasta donde se depositan los objetos en juego, sobre la mesa se gira la pirinola, alrededor de la mesa se encuentran los jugadores. La pirinola se coloca en la mesa frente al siguiente jugador que la hará girar (el que se encuentra a la derecha del último jugador que hizo girar la pirinola).

De acuerdo a lo anterior, las caracterísiticas que definen a la mesa de juego son:
* la pirinola
* los jugadores, por lo menos dos
* la canasta, inicialmente vacia
* el número de túrno del siguiente jugador que girará la pirinola
Las acciones que se puede realizar en la mesa son:
* obtener el turno del jugador que va a girar la pirinola
* avanzar el turno al siguiente jugador
* cuando en la cara de la pirinola que queda en la parte superior es Toma 1 o Toma 2 el jugador en turno debe extraer uno o dos objetos de la canasta y almacenarlos con los objetos propios, y avanzar el turno al siguiente jugador
* cuando en la cara de la pirinola que queda en la parte superior es Pon 1 o Pon 2 el jugador debe extraer uno o dos objetos de los propios y colocarlos en la canasta, y avanzar el turno al siguiente jugador.
* cuando en la cara de la pirinola que queda en la parte superior es todos ponen, cada jugador debe poner un objeto de los propios en la canasta.
* jugar. Una vez definidos todos los elementos se puede jugar el juego sobre la mesa siguiendo las reglas.
    * Todos ponen 5 objetos
    * Mientras la canasta no está vacia
        * determinar cuál es el jugador que le toca girar la pirinola
        * girar la piriniola
        * ejecutar la acción de acuerdo a la cara que se muestra en la parte superior de la pirinola
        * si la canasta esta vacia el jugador que realizó el giro ganó
        * si la canasta no está vacía avanzar el turno al siguiente jugador
***
## Clase mesa
```python
class mesa(object):
    def __init__(self, ponderadores = None, jugadores= None, canasta_ = None):
        self.set_pirinola(ponderadores)
        self.set_jugadores(jugadores)
        self.set_canasta(canasta_)
        self.set_turno(0)
    def set_pirinola(self, ponderadores):
        self.__pirinola = pirinola(ponderadores)
    def get_pirinola(self):
        return self.__pirinola
    def set_jugadores(self, jugadores):
        if jugadores == None:
            self.__jugadores = [jugador() for _ in range(4)]
        else:
            self.__jugadores = [jugador(jugador_) for jugador_ in jugadores]
    def get_jugadores(self):
        return self.__jugadores
    def cantidad_jugadores(self):
        return len(self.get_jugadores())
    def set_canasta(self, canasta_):
        self.__canasta = canasta()
    def get_canasta(self):
        return self.__canasta
    def set_turno(self, turno):
        self.__turno = turno
    def get_turno(self):
        return self.__turno
    def avanza_turno(self):
        self.set_turno( (self.get_turno()+1) % self.cantidad_jugadores() )
    def toma(self,jugador,cuantos):
        self.get_canasta().extraer(cuantos)
        self.get_jugadores()[self.get_turno()].recibir(cuantos)
        self.avanza_turno()
    def pon(self,jugador,cuantos):
        jb = self.get_jugadores()[self.get_turno()].dar(cuantos)        
        self.get_canasta().agregar(jb)
        self.avanza_turno()
    def todos_ponen(self, n):
        for jugador in self.get_jugadores():
            self.pon(jugador,n)
            
    def jugar(self):
        self.todos_ponen(5)
        tiros = 0
        while not self.get_canasta().vacia() and tiros< 15:
            print(f"Hay {self.get_canasta().get_contenido()} jellybeans en la mesa")            
            resultado = self.get_pirinola().girar()
            turno = self.get_turno()
            jugador = self.get_jugadores()[turno]
            if resultado == "Pon 1":
                self.pon(jugador, 1)
            elif resultado == "Pon 2":
                self.pon(jugador, 2)
            elif resultado == "Toma 1":
                self.toma(turno, 1)
            elif resultado == "Toma 2":
                self.toma(turno,2)
            elif resultado == "Toma todo":
                self.toma(turno, self.get_canasta().get_contenido())
                print(f"Ganó el jugador {self.get_jugadores()[turno]}")
            elif resultado == "Todos ponen":
                self.todos_ponen(1)
            tiros += 1
            self.avanza_turno()            
            print(f"{resultado} {jugador}")
            sleep(1.5)
    def __str__(self):
        return f"Hay {self.get_canasta().get_contenido()} jellybeans en la mesa"
```
